```python
#importing the packages needed
import janitor
import pandas as pd
import networkx as nx
from pyvis import network as net
from IPython.core.display import display, HTML
#this visualization came from a turotial about viusualizing your linkedin network. As such the link can be found at the bottom. 
#I found this very very interesting. I was able to download my linkedin connection from linkedin directly. The .csv will be in my .zip file. 
```

```python
df = pd.read_csv('data/Connections.csv', skiprows=2) #reading in my connections as a datafram
df_company = df['Company'].value_counts().reset_index() #filtering out to just company and count
df_company.columns = ['Company', 'count']
df_company = df_company.sort_values(by="count", ascending=False)
df_company #unique 'companies' and then summing them so we can display my linkedin network
```

```python
# initializing the network graph
g = nx.Graph()
g.add_node('Jake (Me)') #this is the root node or me in this case, Jake and then all of my nconnection branch from me. The size of the node represents the number of connections i have there. The bigger the node, the more the connections. 

#intterating though each row of the df.company dataframe and adding them to the network graph. 
for _, row in df_company.iterrows():
    company = row['Company']
    count = row['count']
    title = f"<b>{company}</b> â€“ {count}"
    positions = set([x for x in df[company == df['Company']]['Position']])
    positions = ''.join('<li>{}</li>'.format(x) for x in positions)
    position_list = f"<ul>{positions}</ul>"
    hover_info = title + position_list
    g.add_node(company, size=count*2, title=hover_info, color='#33ff68')
    g.add_edge('Jake (Me)', company, color='white')

# generating the graph
nt = net.Network(height='1100px', width='1100px', bgcolor="black", font_color='white')
display(HTML('linkedin_graph.html'))
nt.from_nx(g)
nt.hrepulsion()
nt.show('linkedin_graph.html')
display(HTML('linkedin_graph.html'))
```
